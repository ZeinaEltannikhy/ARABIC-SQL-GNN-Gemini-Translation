```python
import os
import re
import sqlite3
from tqdm import tqdm
import networkx as nx
import torch
from torch_geometric.data import Data
import matplotlib.pyplot as plt
import random

# Ensure plots display in notebook
%matplotlib inline

# --- Function to Extract Schema from SQL Files with better regex patterns ---
def extract_schema_from_sql(sql_file_path):
    schema = {}
    try:
        with open(sql_file_path, 'r', encoding='utf-8', errors='ignore') as file:
            sql_content = file.read()
            
        # Extract CREATE TABLE statements (more flexible pattern)
        create_table_pattern = r'[cC][rR][eE][aA][tT][eE]\s+[tT][aA][bB][lL][eE]\s+(?:IF\s+NOT\s+EXISTS\s+)?[`"\[]?(\w+)[`"\]]?\s*\(([\s\S]*?)\);'
        tables = re.findall(create_table_pattern, sql_content)
        
        for table_name, table_content in tables:
            columns = []
            foreign_keys = []
            
            # Split by commas but handle parentheses correctly
            lines = []
            current_line = ""
            paren_level = 0
            
            for char in table_content:
                if char == '(':
                    paren_level += 1
                    current_line += char
                elif char == ')':
                    paren_level -= 1
                    current_line += char
                elif char == ',' and paren_level == 0:
                    lines.append(current_line.strip())
                    current_line = ""
                else:
                    current_line += char
            
            if current_line.strip():
                lines.append(current_line.strip())
            
            for line in lines:
                if line and not line.startswith('--'):  # Skip empty lines and comments
                    # Check if line is a foreign key constraint
                    fk_match = re.search(r'FOREIGN\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)\s*REFERENCES\s*[`"\[]?(\w+)[`"\]]?\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)', line, re.IGNORECASE)
                    if fk_match:
                        from_col, to_table, to_col = fk_match.groups()
                        foreign_keys.append({"from_col": from_col, "to_table": to_table, "to_col": to_col})
                    else:
                        # Try to parse as a column definition
                        col_match = re.search(r'^\s*[`"\[]?(\w+)[`"\]]?\s+([\w\s\(\)]+)', line)
                        if col_match:
                            col_name = col_match.group(1)
                            col_type = col_match.group(2).strip()
                            
                            # Check for PRIMARY KEY constraint
                            is_pk = 1 if "PRIMARY KEY" in line.upper() else 0
                            
                            # Check for NOT NULL constraint
                            is_notnull = 1 if "NOT NULL" in line.upper() else 0
                            
                            columns.append({
                                "name": col_name,
                                "type": col_type,
                                "notnull": is_notnull,
                                "default": None,
                                "pk": is_pk
                            })
            
            # Look for separate PRIMARY KEY declarations
            pk_pattern = r'PRIMARY\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)'
            pk_matches = re.findall(pk_pattern, table_content, re.IGNORECASE)
            for pk_col in pk_matches:
                for col in columns:
                    if col["name"] == pk_col:
                        col["pk"] = 1
                        
            # Look for separate FOREIGN KEY declarations that might not be in column format
            fk_pattern = r'FOREIGN\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)\s*REFERENCES\s*[`"\[]?(\w+)[`"\]]?\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)'
            fk_matches = re.findall(fk_pattern, table_content, re.IGNORECASE)
            for from_col, to_table, to_col in fk_matches:
                foreign_keys.append({"from_col": from_col, "to_table": to_table, "to_col": to_col})
            
            if columns:  # Only add if we have columns
                schema[table_name] = {"columns": columns, "foreign_keys": foreign_keys}
                
    except Exception as e:
        print(f"Error processing {sql_file_path}: {e}")
        return None
    
    return schema

# --- Functions for Graph Creation and Visualization ---
def build_schema_graph(schema):
    G = nx.DiGraph()
    for table, info in schema.items():
        G.add_node(table, type="table", feature=[1.0, 0.0, 0.0], name=table)
        primary_keys = [col['name'] for col in info["columns"] if col.get('pk', 0) == 1]
        for col in info["columns"]:
            col_node = f"{table}.{col['name']}"
            G.add_node(col_node, type="column", feature=[0.0, 1.0, 1.0 if col['name'] in primary_keys else 0.0], name=col['name'], table=table)
            G.add_edge(table, col_node, relation="contains")
        for fk in info["foreign_keys"]:
            from_col_node = f"{table}.{fk['from_col']}"
            to_col_node = f"{fk['to_table']}.{fk['to_col']}"
            if G.has_node(from_col_node) and G.has_node(to_col_node):
                G.add_edge(from_col_node, to_col_node, relation="foreign_key")
                G.add_edge(table, fk["to_table"], relation="references")
    print(f"Graph Stats for {list(schema.keys())[0] if schema else 'unknown'} - Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
    return G if G.number_of_nodes() > 0 else None

def graph_to_data(G):
    node_mapping = {n: i for i, n in enumerate(G.nodes)}
    x = torch.tensor([G.nodes[n].get("feature", [1.0, 0.0, 0.0]) for n in G.nodes], dtype=torch.float)
    edge_index = torch.tensor([[node_mapping[u], node_mapping[v]] for u, v in G.edges], dtype=torch.long).t().contiguous()
    edge_attr = torch.tensor([[1,0,0] if d['relation'] == 'contains' else [0,1,0] if d['relation'] == 'foreign_key' else [0,0,1] for _, _, d in G.edges(data=True)], dtype=torch.float)
    return Data(x=x, edge_index=edge_index, edge_attr=edge_attr)

def visualize_graph(G, db_id):
    plt.figure(figsize=(14, 10))
    node_colors = ['#87CEEB' if G.nodes[n]['type'] == 'table' else '#98FB98' for n in G.nodes]
    node_sizes = [2000 if G.nodes[n]['type'] == 'table' else 1000 for n in G.nodes]
    edge_styles = []
    edge_colors = []
    for _, _, data in G.edges(data=True):
        if data['relation'] == 'contains':
            edge_colors.append('#333333')
            edge_styles.append('solid')
        elif data['relation'] == 'foreign_key':
            edge_colors.append('#FF4500')
            edge_styles.append('dashed')
        elif data['relation'] == 'references':
            edge_colors.append('#4682B4')
            edge_styles.append('dotted')
    pos = nx.spring_layout(G, k=1.5, iterations=100, scale=2)
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes,
                           node_shape='s' if G.nodes[list(G.nodes)[0]]['type'] == 'table' else 'o',
                           edgecolors='black', linewidths=1.5)
    for (u, v, d), color, style in zip(G.edges(data=True), edge_colors, edge_styles):
        nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], edge_color=color, style=style,
                               arrows=True, arrowstyle='-|>', arrowsize=25, width=2)
    labels = {node: data['name'] for node, data in G.nodes(data=True)}
    nx.draw_networkx_labels(G, pos, labels, font_size=12, font_weight='bold', font_color='black',
                            bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))
    plt.title(f"Schema Graph: {db_id}", fontsize=16, pad=20)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# --- Main Processing Pipeline (SQL files only) ---
database_dir = "/kaggle/input/txttosql-nlp/database"
schemas = {}
graphs = {}
gnn_data = {}

# Process all databases
db_folders = [f for f in os.listdir(database_dir) if os.path.isdir(os.path.join(database_dir, f))]
for db_folder in tqdm(db_folders, desc="Processing Databases"):
    db_path = os.path.join(database_dir, db_folder)
    schema = None
    found_schema = False
    
    # 1. Try folder-named file first (e.g., "car_1/car_1.sql")
    folder_named_file = os.path.join(db_path, f"{db_folder}.sql")
    if os.path.exists(folder_named_file):
        schema = extract_schema_from_sql(folder_named_file)
        if schema and len(schema) > 0:
            schemas[db_folder] = schema
            print(f"Extracted schema from {db_folder}.sql for {db_folder}: {list(schema.keys())}")
            found_schema = True
    
    # 2. Try common schema filenames if folder-named file didn't work
    if not found_schema:
        common_filenames = ['schema.sql', 'create_tables.sql', 'tables.sql', 'database.sql', 'ddl.sql']
        for filename in common_filenames:
            file_path = os.path.join(db_path, filename)
            if os.path.exists(file_path):
                schema = extract_schema_from_sql(file_path)
                if schema and len(schema) > 0:
                    schemas[db_folder] = schema
                    print(f"Extracted schema from {filename} for {db_folder}: {list(schema.keys())}")
                    found_schema = True
                    break
    
    # 3. If still no schema found, try any SQL file
    if not found_schema:
        sql_files = [f for f in os.listdir(db_path) if f.endswith(".sql")]
        for sql_file in sql_files:
            file_path = os.path.join(db_path, sql_file)
            schema = extract_schema_from_sql(file_path)
            if schema and len(schema) > 0:
                schemas[db_folder] = schema
                print(f"Extracted schema from {sql_file} for {db_folder}: {list(schema.keys())}")
                found_schema = True
                break
    
    # 4. Build graph if schema was found
    if found_schema and schema:
        graph = build_schema_graph(schema)
        if graph:
            graphs[db_folder] = graph
            gnn_data[db_folder] = graph_to_data(graph)
    else:
        print(f"Could not extract schema for {db_folder}")

# Sample databases for visualization (if any)
if graphs:
    sample_db_ids = random.sample(list(graphs.keys()), min(3, len(graphs)))
    print(f"Visualizing graphs for: {sample_db_ids}")
    for db_id in sample_db_ids:
        visualize_graph(graphs[db_id], db_id)
else:
    print("No graphs were created, cannot visualize.")

print(f"Processed {len(schemas)} schemas, {len(graphs)} graphs, and {len(gnn_data)} GNN datasets.")
```

```python
import os
import json
import re
from collections import defaultdict
import sqlite3
import networkx as nx
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
import google.generativeai as genai
from torch_geometric.data import Data
import torch

# Configure Gemini API
os.environ["GOOGLE_API_KEY"] = "AIzaSyCGdniZN23Mh-a50V5NP6SkONzPlWuSYVo"  # Replace with your API key
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
gemini_model = genai.GenerativeModel("gemini-1.5-pro")

# Load dataset
def load_ar_spider_dataset(jsonl_file):
    dataset = []
    with open(jsonl_file, 'r', encoding='utf-8') as f:
        for line in f:
            entry = json.loads(line.strip())
            dataset.append({
                'question': entry['question'],
                'arabic': entry['arabic'],
                'query': entry['query'],
                'db_id': entry['db_id']
            })
    return dataset

# Group dataset by db_id
ar_spider_dataset = load_ar_spider_dataset('/kaggle/input/txttosql-nlp/AR_spider.jsonl')
dataset_by_db = defaultdict(list)
for entry in ar_spider_dataset:
    dataset_by_db[entry['db_id']].append(entry)

# Assume schemas is already populated from your original code
# Example: schemas = extract_schemas_from_directory('Dataset/database')

# Build retriever
retriever = SentenceTransformer('distiluse-base-multilingual-cased-v2')

def build_retriever(dataset, schemas):
    texts = []
    metadata = []
    for db_id, schema in schemas.items():
        for table, info in schema.items():
            table_text = f"Table: {table}, Columns: {', '.join(col['name'] for col in info['columns'])}"
            texts.append(table_text)
            metadata.append({'type': 'schema', 'db_id': db_id, 'table': table})
        for entry in dataset_by_db.get(db_id, []):
            texts.append(entry['arabic'])
            metadata.append({'type': 'example', 'db_id': db_id, 'sql': entry['query']})
    
    embeddings = retriever.encode(texts, convert_to_numpy=True)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)
    return index, texts, metadata

index, texts, metadata = build_retriever(ar_spider_dataset, schemas)

def retrieve(query, db_id, index, texts, metadata, k=3):
    query_embedding = retriever.encode([query], convert_to_numpy=True)
    distances, indices = index.search(query_embedding, k * 2)
    results = [(texts[i], metadata[i]) for i in indices[0] if metadata[i]['db_id'] == db_id]
    return results[:k]

# SQL generation with Gemini
def format_input(query, retrieved_items, max_context=2):
    context = []
    for text, meta in retrieved_items[:max_context]:
        if meta['type'] == 'example' and meta['db_id'] == 'department_management':
            context.append(f"Example Query: {text}\nExample SQL: {meta['sql']}")
        elif meta['type'] == 'schema':
            context.append(f"Schema: {text}")
    context_str = "\n\n".join(context)
    return f"""
Task: Generate an SQL query for the given Arabic question based on the provided schema and example.

Arabic Question: {query}

Context:
{context_str}

Instructions:
- Generate only the SQL query.
- Use the schema and example to ensure correctness.
- Do not include explanations or additional text.
"""

def generate_sql(query, retrieved_items):
    input_prompt = format_input(query, retrieved_items)
    response = gemini_model.generate_content(
        input_prompt,
        generation_config=genai.types.GenerationConfig(
            temperature=0.2,
            top_p=0.9,
            max_output_tokens=128
        )
    )
    return response.text.strip()

# Execution evaluation
def evaluate_sql(generated_sql, expected_sql, db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    try:
        cursor.execute(generated_sql)
        gen_results = cursor.fetchall()
        cursor.execute(expected_sql)
        exp_results = cursor.fetchall()
        return gen_results == exp_results
    except Exception as e:
        print(f"Execution error: {e}")
        return False
    finally:
        conn.close()

# Full pipeline
def arabic_to_sql_pipeline(query, db_id, database_dir, expected_sql=None):
    # Retrieve context
    results = retrieve(query, db_id, index, texts, metadata, k=3)
    print("Retrieved items:")
    for text, meta in results:
        print(f"- {text}, Metadata: {meta}")
    
    # Generate SQL
    sql = generate_sql(query, results)
    print(f"Generated SQL: {sql}")
    
    # Validate
    if expected_sql:
        print(f"Expected SQL: {expected_sql}")
        exact_match = sql.strip() == expected_sql.strip()
        print(f"Exact Match: {exact_match}")
        
        # Execution evaluation
        db_path = os.path.join(database_dir, db_id, f"{db_id}.sqlite")
        if os.path.exists(db_path):
            execution_match = evaluate_sql(sql, expected_sql, db_path)
            print(f"Execution Match: {execution_match}")
        else:
            print(f"Database not found: {db_path}")
    
    return sql

# Test the pipeline
query = "كم عدد رؤساء الأقسام الذين تزيد أعمارهم عن 56 سنة؟"
db_id = "department_management"
database_dir = "Dataset/database"
expected_sql = "SELECT count(*) FROM head WHERE age > 56"
sql = arabic_to_sql_pipeline(query, db_id, database_dir, expected_sql)
```

```python
import os
import json
import re
from collections import defaultdict
import sqlite3
import networkx as nx
import torch
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
import google.generativeai as genai
from tqdm import tqdm
import random
import matplotlib.pyplot as plt

# Configure Gemini API
os.environ["GOOGLE_API_KEY"] = "AIzaSyCGdniZN23Mh-a50V5NP6SkONzPlWuSYVo"  # Replace with your API key
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
gemini_model = genai.GenerativeModel("gemini-1.5-pro")

# --- Schema Extraction and Graph Creation ---
def extract_schema_from_sql(sql_file_path):
    schema = {}
    try:
        with open(sql_file_path, 'r', encoding='utf-8', errors='ignore') as file:
            sql_content = file.read()
        create_table_pattern = r'[cC][rR][eE][aA][tT][eE]\s+[tT][aA][bB][lL][eE]\s+(?:IF\s+NOT\s+EXISTS\s+)?[`"\[]?(\w+)[`"\]]?\s*\(([\s\S]*?)\);'
        tables = re.findall(create_table_pattern, sql_content)
        for table_name, table_content in tables:
            columns = []
            foreign_keys = []
            lines = []
            current_line = ""
            paren_level = 0
            for char in table_content:
                if char == '(':
                    paren_level += 1
                    current_line += char
                elif char == ')':
                    paren_level -= 1
                    current_line += char
                elif char == ',' and paren_level == 0:
                    lines.append(current_line.strip())
                    current_line = ""
                else:
                    current_line += char
            if current_line.strip():
                lines.append(current_line.strip())
            for line in lines:
                if line and not line.startswith('--'):
                    fk_match = re.search(r'FOREIGN\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)\s*REFERENCES\s*[`"\[]?(\w+)[`"\]]?\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)', line, re.IGNORECASE)
                    if fk_match:
                        from_col, to_table, to_col = fk_match.groups()
                        foreign_keys.append({"from_col": from_col, "to_table": to_table, "to_col": to_col})
                    else:
                        col_match = re.search(r'^\s*[`"\[]?(\w+)[`"\]]?\s+([\w\s\(\)]+)', line)
                        if col_match:
                            col_name = col_match.group(1)
                            col_type = col_match.group(2).strip()
                            is_pk = 1 if "PRIMARY KEY" in line.upper() else 0
                            is_notnull = 1 if "NOT NULL" in line.upper() else 0
                            columns.append({
                                "name": col_name,
                                "type": col_type,
                                "notnull": is_notnull,
                                "default": None,
                                "pk": is_pk
                            })
            pk_pattern = r'PRIMARY\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)'
            pk_matches = re.findall(pk_pattern, table_content, re.IGNORECASE)
            for pk_col in pk_matches:
                for col in columns:
                    if col["name"] == pk_col:
                        col["pk"] = 1
            fk_pattern = r'FOREIGN\s+KEY\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)\s*REFERENCES\s*[`"\[]?(\w+)[`"\]]?\s*\(\s*[`"\[]?(\w+)[`"\]]?\s*\)'
            fk_matches = re.findall(fk_pattern, table_content, re.IGNORECASE)
            for from_col, to_table, to_col in fk_matches:
                foreign_keys.append({"from_col": from_col, "to_table": to_table, "to_col": to_col})
            if columns:
                schema[table_name] = {"columns": columns, "foreign_keys": foreign_keys}
    except Exception as e:
        print(f"Error processing {sql_file_path}: {e}")
        return None
    return schema

def build_schema_graph(schema):
    G = nx.DiGraph()
    for table, info in schema.items():
        G.add_node(table, type="table", feature=[1.0, 0.0, 0.0], name=table)
        primary_keys = [col['name'] for col in info["columns"] if col.get('pk', 0) == 1]
        for col in info["columns"]:
            col_node = f"{table}.{col['name']}"
            G.add_node(col_node, type="column", feature=[0.0, 1.0, 1.0 if col['name'] in primary_keys else 0.0], name=col['name'], table=table)
            G.add_edge(table, col_node, relation="contains")
        for fk in info["foreign_keys"]:
            from_col_node = f"{table}.{fk['from_col']}"
            to_col_node = f"{fk['to_table']}.{fk['to_col']}"
            if G.has_node(from_col_node) and G.has_node(to_col_node):
                G.add_edge(from_col_node, to_col_node, relation="foreign_key")
                G.add_edge(table, fk["to_table"], relation="references")
    print(f"Graph Stats for {list(schema.keys())[0] if schema else 'unknown'} - Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
    return G if G.number_of_nodes() > 0 else None

def graph_to_data(G):
    node_mapping = {n: i for i, n in enumerate(G.nodes)}
    x = torch.tensor([G.nodes[n].get("feature", [1.0, 0.0, 0.0]) for n in G.nodes], dtype=torch.float)
    edge_index = torch.tensor([[node_mapping[u], node_mapping[v]] for u, v in G.edges], dtype=torch.long).t().contiguous()
    edge_attr = torch.tensor([[1,0,0] if d['relation'] == 'contains' else [0,1,0] if d['relation'] == 'foreign_key' else [0,0,1] for _, _, d in G.edges(data=True)], dtype=torch.float)
    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr)
    data.node_mapping = node_mapping
    return data

# Process schemas and graphs
database_dir = "/kaggle/input/txttosql-nlp/database"
schemas = {}
graphs = {}
gnn_data = {}

db_folders = [f for f in os.listdir(database_dir) if os.path.isdir(os.path.join(database_dir, f))]
for db_folder in tqdm(db_folders, desc="Processing Databases"):
    db_path = os.path.join(database_dir, db_folder)
    schema = None
    found_schema = False
    folder_named_file = os.path.join(db_path, f"{db_folder}.sql")
    if os.path.exists(folder_named_file):
        schema = extract_schema_from_sql(folder_named_file)
        if schema and len(schema) > 0:
            schemas[db_folder] = schema
            found_schema = True
    if not found_schema:
        common_filenames = ['schema.sql', 'create_tables.sql', 'tables.sql', 'database.sql', 'ddl.sql']
        for filename in common_filenames:
            file_path = os.path.join(db_path, filename)
            if os.path.exists(file_path):
                schema = extract_schema_from_sql(file_path)
                if schema and len(schema) > 0:
                    schemas[db_folder] = schema
                    found_schema = True
                    break
    if not found_schema:
        sql_files = [f for f in os.listdir(db_path) if f.endswith(".sql")]
        for sql_file in sql_files:
            file_path = os.path.join(db_path, sql_file)
            schema = extract_schema_from_sql(file_path)
            if schema and len(schema) > 0:
                schemas[db_folder] = schema
                found_schema = True
                break
    if found_schema and schema:
        graph = build_schema_graph(schema)
        if graph:
            graphs[db_folder] = graph
            gnn_data[db_folder] = graph_to_data(graph)

# --- GNN Model ---
class SchemaGNN(torch.nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, output_dim=1):
        super(SchemaGNN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)
    
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Mock GNN prediction
def predict_relevant_schema_elements(query, db_id, gnn_data, schemas):
    relevant_nodes = []
    schema = schemas.get(db_id, {})
    for table, info in schema.items():
        relevant_nodes.append(table)
        for col in info["columns"]:
            relevant_nodes.append(f"{table}.{col['name']}")
    return relevant_nodes[:5]

# --- Dataset and Retriever ---
def load_ar_spider_dataset(jsonl_file):
    dataset = []
    with open(jsonl_file, 'r', encoding='utf-8') as f:
        for line in f:
            entry = json.loads(line.strip())
            dataset.append({
                'question': entry['question'],
                'arabic': entry['arabic'],
                'query': entry['query'],
                'db_id': entry['db_id']
            })
    return dataset

ar_spider_dataset = load_ar_spider_dataset('/kaggle/input/txttosql-nlp/AR_spider.jsonl')
dataset_by_db = defaultdict(list)
for entry in ar_spider_dataset:
    dataset_by_db[entry['db_id']].append(entry)

retriever = SentenceTransformer('distiluse-base-multilingual-cased-v2')

def build_retriever(dataset, schemas):
    texts = []
    metadata = []
    for db_id, schema in schemas.items():
        for table, info in schema.items():
            table_text = f"Table: {table}, Columns: {', '.join(col['name'] for col in info['columns'])}"
            texts.append(table_text)
            metadata.append({'type': 'schema', 'db_id': db_id, 'table': table})
        for entry in dataset_by_db.get(db_id, []):
            texts.append(entry['arabic'])
            metadata.append({'type': 'example', 'db_id': db_id, 'sql': entry['query']})
    embeddings = retriever.encode(texts, convert_to_numpy=True)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)
    return index, texts, metadata

index, texts, metadata = build_retriever(ar_spider_dataset, schemas)

def retrieve(query, db_id, index, texts, metadata, k=3):
    query_embedding = retriever.encode([query], convert_to_numpy=True)
    distances, indices = index.search(query_embedding, k * 2)
    results = [(texts[i], metadata[i]) for i in indices[0] if metadata[i]['db_id'] == db_id]
    return results[:k]

# --- SQL Generation with Gemini ---
def format_input(query, retrieved_items, gnn_elements, max_context=2):
    context = []
    for text, meta in retrieved_items[:max_context]:
        if meta['type'] == 'example' and meta['db_id'] == 'department_management':
            context.append(f"Example Query: {text}\nExample SQL: {meta['sql']}")
        elif meta['type'] == 'schema':
            context.append(f"Schema: {text}")
    context.append(f"GNN Predicted Schema Elements: {', '.join(gnn_elements)}")
    context_str = "\n\n".join(context)
    return f"""
Task: Generate an SQL query for the given Arabic question based on the provided schema, example, and GNN-predicted schema elements.

Arabic Question: {query}

Context:
{context_str}

Instructions:
- Generate only the SQL query.
- Use the schema, example, and GNN predictions to ensure correctness.
- Do not include explanations or additional text.
"""

def generate_sql(query, retrieved_items, gnn_elements):
    input_prompt = format_input(query, retrieved_items, gnn_elements)
    response = gemini_model.generate_content(
        input_prompt,
        generation_config=genai.types.GenerationConfig(
            temperature=0.2,
            top_p=0.9,
            max_output_tokens=128
        )
    )
    return response.text.strip()

# --- Execution Evaluation ---
def evaluate_sql(generated_sql, expected_sql, db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    try:
        cursor.execute(generated_sql)
        gen_results = cursor.fetchall()
        cursor.execute(expected_sql)
        exp_results = cursor.fetchall()
        return gen_results == exp_results
    except Exception as e:
        print(f"Execution error: {e}")
        return False
    finally:
        conn.close()

# --- Full Pipeline ---
def arabic_to_sql_pipeline(query, db_id, database_dir, schemas, gnn_data, expected_sql=None):
    gnn_elements = predict_relevant_schema_elements(query, db_id, gnn_data, schemas)
    print("GNN Predicted Schema Elements:", gnn_elements)
    results = retrieve(query, db_id, index, texts, metadata, k=3)
    print("Retrieved items:")
    for text, meta in results:
        print(f"- {text}, Metadata: {meta}")
    sql = generate_sql(query, results, gnn_elements)
    print(f"Generated SQL: {sql}")
    if expected_sql:
        print(f"Expected SQL: {expected_sql}")
        exact_match = sql.strip() == expected_sql.strip()
        print(f"Exact Match: {exact_match}")
        db_path = os.path.join(database_dir, db_id, f"{db_id}.sqlite")
        if os.path.exists(db_path):
            execution_match = evaluate_sql(sql, expected_sql, db_path)
            print(f"Execution Match: {execution_match}")
        else:
            print(f"Database not found: {db_path}")
    return sql

# Test the pipeline
query = "كم عدد رؤساء الأقسام الذين تزيد أعمارهم عن 56 سنة؟"
db_id = "department_management"
database_dir = "Dataset/database"
expected_sql = "SELECT count(*) FROM head WHERE age > 56"
sql = arabic_to_sql_pipeline(query, db_id, database_dir, schemas, gnn_data, expected_sql)
```
